```python
import pandas as pd
import ast

# 5. Parse list/dict columns
for col in ['features','description','images','videos','categories','details']:
    # turn the stringified list/dict into an actual Python object
    meta_df[col] = meta_df[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# 6. Feature engineering & extraction

# 6a) Count how many features each product has
meta_df['n_features'] = meta_df['features'].apply(len)

# 6b) Length of description list
meta_df['n_description_items'] = meta_df['description'].apply(len)

# 6c) Grab the first image URL (thumb) if any
def first_image_thumb(img_list):
    if img_list and isinstance(img_list, list):
        return img_list[0].get('thumb')
    return None
meta_df['first_image'] = meta_df['images'].apply(first_image_thumb)

# 6d) Extract ‘brand’ and ‘color’ and release date from details dict
def extract_detail(key):
    return meta_df['details'].apply(lambda d: d.get(key) if isinstance(d, dict) else None)

meta_df['brand'] = extract_detail('Brand')
meta_df['color'] = extract_detail('Color')
meta_df['date_first_available'] = extract_detail('Date First Available')

# 6e) Simplify categories to its first element
meta_df['primary_category'] = meta_df['categories'].apply(lambda lst: lst[0] if isinstance(lst, list) and lst else None)

# 7. Drop columns no longer needed for graph
drop_cols = ['features','description','images','videos','categories','details','bought_together','subtitle','author']
meta_df.drop(columns=[c for c in drop_cols if c in meta_df], inplace=True)

# 8. Final head & missing check
print("----- Cleaned DataFrame Head -----")
print(meta_df.head(5))

print("\n----- Remaining Missing Values -----")
print(meta_df.isnull().sum())
```

```python
meta_df.head(30)
```

```python
import pandas as pd
import numpy as np


# 1) Parse `date_first_available` into a true datetime, drop if it fails
meta_df['date_first_available'] = pd.to_datetime(
    meta_df['date_first_available'],
    errors='coerce',
    format='%B %d, %Y'
)
meta_df.dropna(subset=['date_first_available'], inplace=True)

# 2) Fill missing brands and colors with a sentinel
meta_df['brand'] = meta_df['brand'].fillna('Unknown')
meta_df['color'] = meta_df['color'].fillna('Unknown')

# 3) Remove products with zero features _and_ zero description items (likely bad records)
mask = ~((meta_df['n_features'] == 0) & (meta_df['n_description_items'] == 0))
meta_df = meta_df[mask]

# 4) Clip out extreme prices (e.g. > 99th percentile) to reduce outlier skew
p99 = meta_df['price'].quantile(0.99)
meta_df = meta_df[meta_df['price'] <= p99]

# 5) Create a simple rating bin (Low/Med/High)
meta_df['rating_bin'] = pd.cut(
    meta_df['average_rating'],
    bins=[0, 3.5, 4.5, 5.0],
    labels=['Low','Medium','High']
)

# 6) Reset index for cleanliness
meta_df.reset_index(drop=True, inplace=True)

# 7) Final sanity checks
print("----- Final shape:", meta_df.shape)
print("\n----- Missing values per column -----")
print(meta_df.isnull().sum())
print("\n----- Price summary after clipping -----")
print(meta_df['price'].describe())
print("\n----- Rating bins distribution -----")
print(meta_df['rating_bin'].value_counts())
```

```python
# 1) Fill missing primary_category with main_category
meta_df['primary_category'] = (
    meta_df['primary_category']
    .fillna(meta_df['main_category'])
)

# 2) Drop any rows still missing after that (should be none)
meta_df.dropna(subset=['primary_category'], inplace=True)

# 3) Normalize text columns (strip whitespace, unify case)
for col in ['main_category','store','brand','color','rating_bin']:
    meta_df[col] = (
        meta_df[col]
        .astype(str)
        .str.strip()
        .str.title()
    )

# 4) Clean up funky brand values (e.g. remove leading “***”)
meta_df['brand'] = meta_df['brand'].str.replace(r'^\*+','',regex=True).replace('Unknown','Unknown')

# 5) Final sanity checks
print("Final shape:", meta_df.shape)
print("\nMissing values by column:\n", meta_df.isnull().sum())
print("\nUnique primary_category values:", meta_df['primary_category'].unique())
print("\nSample:\n", meta_df.head(3))
```
